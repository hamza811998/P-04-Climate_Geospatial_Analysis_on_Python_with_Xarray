# 02. Select and filter data through coordinates

Xarray selectors allows you to filter across multidimensions without hassles, and it is an very powerful feature when dealing with geospatial data.

```python
ds.sel(longitude=-82)
```

```python
ds.sel(longitude=-82, latitude=5.75, time='2018-01-01')
```

```python
ds.sel(latitude=99, method='nearest')
```

```python
ds.sel(latitude=[-16, -15], longitude=[-46, -47], method='nearest')
```

```python
ds.isel(latitude=0, longitude=0)
```

```python
QUERY = ds.longitude < -64
QUERY = QUERY & (ds.latitude > -8)

ds.where(QUERY, drop=True)
```

# 03. Visualize multidimensional and geospatial variables

```python
ds = xr.open_dataset('data.nc').sel(expver=1)
ds
```

```python
ds.skt.plot()
```

```python
ds.isel(time=0).skt.plot()
```

```python
ds.isel(time=0).lai_hv.plot()
```

```python
ds.isel(longitude=0, latitude=0).skt.plot()
```

```python
ds.isel(time=0).sel(longitude=-59, method='nearest').skt.plot()
```

```python
ds.isel(time=0).sel(longitude=-69, method='nearest').skt.plot()
```

# 04. Apply simple operations over multidimensional data

```python
temp_change = (ds.skt.sel(time='2019-01-01') - ds.skt.sel(time='1979-01-01'))

temp_change
```

```python
temp_change.plot()
```

```python
sazonal_change = (ds.lai_hv.sel(time='2019-07-01') - ds.lai_hv.sel(time='2019-01-01'))
sazonal_change.plot()
```

```python
mean_temp = ds.skt.mean(dim='time')
mean_temp
```

```python
mean_temp.plot()
```

```python
ds.skt.mean(dim=['time', 'longitude']).plot()
```

```python
ds.skt.std(dim='time').plot()
```

```python
ds.skt.median(dim=['latitude', 'longitude']).plot()
```

# 05. Apply grouped operations over multidimensional data

```python
(ds.groupby(ds.time.dt.year).mean(dim=['time', 'latitude', 'longitude']).skt.plot())
```

```python
(ds.groupby(ds.time.dt.year).mean(dim=['time', 'latitude', 'longitude']).skt)
```

```python
(ds.groupby(ds.time.dt.month).mean(dim=['time', 'latitude', 'longitude']).skt.plot())
```

```python
(ds.groupby(ds.time.dt.month).mean(dim=['time', 'latitude', 'longitude']).lai_hv.plot())
```

```python
(ds.resample({'time': '10Y'}).mean(dim=['time', 'latitude', 'longitude']).skt.plot())
```

# 06. Merge and concatenate datasets

```python
da = ds.time.dt.year
da
```

```python
ds_1 = ds.merge(da)
ds_1
```

```python
da = ds.longitude ** 2 + ds.latitude ** 2
da.name = 'squared_distance'
ds_2 = ds.merge(da)
ds_2
```

```python
ds_3 = ds_1.merge(ds_2)
ds_3
```

```python
ds_square_root = ds_3 ** (1/2)
ds_cube_root = ds_3 ** (1/3)
ds_4 = xr.concat([ds_3, ds_square_root, ds_cube_root], 'dim_power')
```

```python
ds_4
```

```python
ds.sel(time='2019-01-01').tp.plot()
```

```python
ds_4.sel(dim_power=0,time='2019-01-01').tp.plot()
```

```python
ds_4.sel(dim_power=1,time='2019-01-01').tp.plot()
```

```python
ds_4.sel(dim_power=2,time='2019-01-01').tp.plot()
```
